## 2.py-translator
 - [Read the Docs](http://pythonhosted.org/py-translate/)
 - #### <span style="color:green"> Free Google Translate API</span>
 - #### <span style="color:green"> Cli interface</span>
 - #### <span style="color:green"> Support proxy</span>
 - #### <span style="color:green">Romanified Transliteration</span>
 - #### <span style="color:Red"> api requires internet to operate.</span>


```python
from py_translator import Translator as gt
```

```python
help(gt.translate)
```

```python
print("\nFRANCH\n")
for k,v in frenchDict.items():
    s = gt().translate(text=k, dest='en').text
    print("given:  "+k+"\noutput: "+s + "\nexpect: " + v + "\n-----------")
```

```python
print("\nGERMAN\n")
for k,v in germanDict.items():
    s = gt().translate(text=k, dest='en').text
    print("given:  "+k+"\noutput: "+s + "\nexpect: " + v + "\n-----------")
```

```python
print("\nMANDARIN\n")
for k,v in mandarinDict.items():
    s = gt().translate(text=k, dest='en').text
    print("given:  "+k+"\noutput: "+s + "\nexpect: " + v + "\n-----------")
```
